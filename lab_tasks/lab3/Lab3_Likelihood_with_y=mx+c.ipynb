```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
import random
import math
x = np.linspace(1,10,10)
y = 5*x + 1
c=math.sqrt(2*math.pi*1)
d=1/c
X=[]
Y=[]
prdt=1
for i in range(len(x)):
    eo = (x[i] - y[i])**2
    prdt = prdt*d*math.exp(-1/(2*1)*eo)
    
    Y.append(prdt)
    
plt.plot(x,Y)
plt.show()
```

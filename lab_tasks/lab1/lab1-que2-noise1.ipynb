```python
import random
```

```python
w0 = np.linspace(-10,10,100)
w1 = np.linspace(-10,10,100) 
#x = [1.63612325,  3.08868428,  3.97857531,  5.64951987,  7.00034091,
#        7.76259277,  8.50317507,  8.90879626, 11.40066286, 10.72984941]

#y = [1.63612325,  3.08868428,  3.97857531,  5.64951987,  7.00034091,
#        7.76259277,  8.50317507,  8.90879626, 11.40066286, 10.72984941]
print(w0)
print(w1)
```

```python
x = np.linspace(-6,6,10)
```

```python
random.seed(1)
noise = np.random.randn(10)
```

```python
y = 5*x + 1
plt.plot(x,y)
yn = 5*x + 1 + noise
plt.scatter(x, yn, color='red');
print(yn)
```

```python
ww0, ww1 = np.meshgrid(w0, w1)

#print(Y)
ww1
```

```python
def f(ww0,ww1,x,y,m):
    e = 0
    for i in range(m):
        e += ((ww0+ww1*x[i]) - yn[i] )**2
    return e
```

```python
Z = f(ww0,ww1,x,y,m=len(x))
print(Z)
```

```python
fig = plt.figure()
#ax = plt.axes(projection='3d')
plt.contour(ww0, ww1, Z, levels = [i for i in np.arange(0,1000,40)])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(ww0, ww1, Z, cmap='autumn', rstride=1, cstride=1, edgecolor='none')
```

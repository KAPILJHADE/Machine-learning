# Likelihood graph for one point(x=5)

```python
import math
u=np.arange(1,10,0.1)
x=5
c=math.sqrt(2*math.pi*1)
d=1/c
X=[]
Y=[]
for i in u:
    eo = (x - i)**2
    prdt = d*math.exp(-1/(2*1)*eo)
    X.append(i)
    Y.append(prdt)
    
plt.plot(X,Y)
plt.show()
```

# Likelihood graph for two points(x=5,7)

```python
import math
u=np.arange(1,10,0.1)
x=[5,7]
c=math.sqrt(2*math.pi*1)
d=1/c
X=[]
Y=[]
for i in u:
    prdt = 0
    for j in range(len(x)):
        eo = (x[j] - i)**2
        prdt = prdt+np.log(d*math.exp(-1/(2*1)*eo))
    
    Y.append(prdt)
    
plt.plot(u,Y)
plt.show()
```

# Likelihood function for two points without scaling(no log)

```python
import math
u=np.arange(1,10,0.1)
x=[5,7]
c=math.sqrt(2*math.pi*1)
d=1/c
X=[]
Y=[]
for i in u:
    prdt = 1
    for j in range(len(x)):
        eo = (x[j] - i)**2
        prdt = prdt*d*math.exp(-1/(2*1)*eo)
    
    Y.append(prdt)
    
plt.plot(u,Y)
plt.show()
```
